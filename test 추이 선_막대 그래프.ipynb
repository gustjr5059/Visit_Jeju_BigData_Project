```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

#   
fm._get_fontconfig_fonts()
font_location = '/Users/user/Library/Fonts/D2Coding-Ver1.3.2-20180524-all.ttc'
font_name = fm.FontProperties(fname=font_location).get_name()
matplotlib.rc('font', family=font_name)

#     .   .
data = pd.read_csv('  2020~2022.csv')

#   
plt.figure(figsize=(10, 8))
plt.bar(data[''], data[''])

#    
mean_val = data[''].mean()
median_val = data[''].median()

#    
plt.axhline(y=mean_val, color='r', linestyle='-', label=f': {mean_val:.2f}')
plt.axhline(y=median_val, color='g', linestyle='-', label=f': {median_val:.2f}')

# ,    
plt.xlabel('')
plt.ylabel('')
plt.title('2020~22    ()')
plt.xticks(rotation=90)
plt.legend()

#  
plt.show()

#    
mean_val, median_val

```

```python
#    .    .

plt.figure(figsize=(10, 8))
plt.plot(data[''], data[''], marker='o')  #   ,    

#    
plt.axhline(y=mean_val, color='r', linestyle='-', label=f': {mean_val:.2f}')
plt.axhline(y=median_val, color='g', linestyle='-', label=f': {median_val:.2f}')

# ,    
plt.xlabel('')
plt.ylabel('')
plt.title('   ( )')
plt.xticks(rotation=90)
plt.legend()

#  
plt.show()
```
